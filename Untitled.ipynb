```python
df
```

```python
df['Education_Level'] = df['Education_Level'].map({"High School":1, "Bachelor's Degree":2, "Master's Degree":3, "PhD": 4})
```

```python
df['Gender'] = df['Gender'].map({'Male':1, 'Female':0})
```

```python
df = pd.get_dummies(df, ['Job_Title'])
```

```python
df
```

```python
x = df.drop('Salary', axis=1) 
y = df['Salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = RobustScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
models = {
    "LR": LinearRegression(),
    "Naive": GaussianNB(),
    "KNN": KNeighborsRegressor(),
    "DT": DecisionTreeRegressor(),
    "RF": RandomForestRegressor(),
    "XGBR": XGBRegressor(),
    "SVR": SVR()
}
```

```python
for name, model in models.items():
    print('-'*30)
    print(f'Using model: {name}')
    
    model.fit(x_train, y_train)
    print(f'Training Score: {model.score(x_train, y_train)}')
    
    y_pred = model.predict(x_test)
    print(f'Test Score: {r2_score(y_test, y_pred)}')  
```

```python
model = RandomForestRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print(f'Training Score: {model.score(x_train, y_train)}')    
print(f'Test Score: {r2_score(y_test, y_pred)}')
```

```python
prediction= pd.DataFrame({'Actual Salary': y_test, 'Predicted Salary': y_pred}).head(20)
prediction
```
